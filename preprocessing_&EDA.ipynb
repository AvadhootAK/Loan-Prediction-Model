```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["Married"].value_counts()
```

```python
df["Married"].fillna(df["Married"].mode()[0], inplace = True)
```

```python
df["Married"].isnull().sum()
```

3 entries were missing (Null) in the married column so they were replced with the 'mode' which is "yes"

```python
df["Dependents"].value_counts()
```

```python
df["Dependents"].fillna(df["Dependents"].mode()[0], inplace = True)
```

```python
df["Dependents"].isnull().sum()
```

Here dependents column had 15 null entries so they were replced with the mode that is 0


```python
df["Gender"].value_counts()
```

```python
df["Gender"].fillna(df["Gender"].mode()[0], inplace = True)
```

```python
df["Gender"].isnull().sum()
```

Gender column had 13 null entires so they were agin replcaed by mode that is Male

```python
df["Self_Employed"].value_counts()
```

```python
df["Self_Employed"].fillna(df["Self_Employed"].mode()[0], inplace= True)
```

32 null entries were present in Self_employed column so they were replaced with the mode that is no

```python
sns.histplot(df['LoanAmount'], kde=True)
```

“LoanAmount distribution was right-skewed (most loans are small, few large ones). Hence, missing values (22) were imputed using the median to reduce outlier influence.”

```python
df["LoanAmount"].fillna(df["LoanAmount"].median(), inplace= True)
```

```python
df["Loan_Amount_Term"].value_counts()
```

Loan Amount term column has 14 null values and after examining the ataset we can see most of the people have term of 360 months that is 30 yrs so we replace the null values with the mode that is 360 months

```python
df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mode()[0], inplace = True)
```

```python
df["Credit_History"].value_counts()
```

Credit history is the most important factore affecting the predictions and this column had 50 null columns

```python
df["Credit_History"].fillna(df["Credit_History"].mode()[0], inplace = True)
```
